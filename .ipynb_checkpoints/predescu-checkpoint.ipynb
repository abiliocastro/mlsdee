```python
# importing libraries

import keras
import numpy as np
from numpy import loadtxt
from numpy import reshape
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras import metrics
from keras import backend as K
from keras.utils.vis_utils import plot_model
from matplotlib import pyplot
```

```python
print('keras: %s' % keras.__version__)
```

```python
# custom R2-score metrics for keras backend
def r2_keras(y_true, y_pred):
    SS_res =  K.sum(K.square(y_true - y_pred)) 
    SS_tot = K.sum(K.square(y_true - K.mean(y_true))) 
    return ( 1 - SS_res/(SS_tot + K.epsilon()) )
```

```python
# custom Magnitude of Relative Error metrics
def mre(y_actual, y_pred):
    res =  K.abs(y_actual - y_pred) / y_actual  
    return res
```

```python
# setting numpy print option
np.set_printoptions(suppress=True)
```

```python
# load the dataset
dataset = loadtxt('desharnais.csv', delimiter=',')
print(dataset.shape)
```

```python
# spliting train and test datasets
# defining split point
split = int(round(dataset.shape[0] * 0.8))
print('SPLIT POINT: %d' % split)
# separating data
train,test = dataset[:split,:],dataset[split:,:]
print('TRAIN SHAPE: %s' % (train.shape,))
print('TEST SHAPE: %s' % (test.shape,))
#print(test)
```

```python
# selecting input and output values
# train
y_train = train[:,5]
print('TRAIN OUTPUT')
print(y_train)
X_train = np.delete(train, [0,3,5], axis=1)
print('TRAIN INPUT')
print(X_train)

# test
y_test = test[:,5]
print('TEST OUTPUT')
print(y_test)
X_test = np.delete(test, [0,3,5], axis=1)
print('TEST INPUT')
print(X_test)
```

```python
# define nn like article
model = Sequential()
model.add(Dense(64, input_dim=9, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='linear'))
```

```python
# compile the keras model
model.compile(loss='mean_absolute_error', optimizer='adam', metrics=[metrics.mse, metrics.mae, r2_keras])
```

```python
# plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
```

```python
# fit the keras model on the dataset
# and
# evaluate the keras model

# model.fit(X_train, y_train, epochs=1000, batch_size=10, validation_split=0.2, verbose=0)
# _, accuracy = model.evaluate(X_train, y_train)
# print('Accuracy: %.2f' % (accuracy*100))

# fit model
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=1000, verbose=0)
```

```python
for i in history.history:
    print(i)
```

```python
# evaluate the model
_,train_mse,train_mae,train_r2 = model.evaluate(X_train, y_train, verbose=0)
_,test_mse,test_mae,test_r2 = model.evaluate(X_test, y_test, verbose=0)
print('TRAIN | mse: %.3f, mae: %.3f, r2: %.3f' % (train_mse, train_mae, train_r2))
print('TEST  | mse: %.3f, mae: %.3f, r2: %.3f' % (test_mse, test_mae, test_r2))
```

```python
# making predictions
predicted = model.predict(X_test)
i = 0
for predic in predicted:
    actual = y_test[i]
    i+=1
    mre =  abs(actual - predic[0]) / actual
    print('AC: %.2f PR: %.2f MRE: %.2f' % (actual,predic,mre))
```

```python
# plot mae
# pyplot.subplot(212)
pyplot.title('Mean Absolute Error')
pyplot.plot(history.history['mean_absolute_error'], label='train')
pyplot.plot(history.history['val_mean_absolute_error'], label='test')
pyplot.legend()
pyplot.show()
```

```python
# plot r2 score
# pyplot.subplot(212)
pyplot.title('R2 Score')
pyplot.plot(history.history['r2_keras'], label='train')
pyplot.plot(history.history['val_r2_keras'], label='test')
pyplot.legend()
pyplot.show()
```
